```python
# Exercice 1

# initialaisation de la variable c

c = "X44bf38j23jdjgfjh737nei47"
```

```python
# programme qui construit deux chaînes de caractères c_alpha et c_num

c_alpha, c_num = "", ""
for i in c:
    if i.isalpha():
        c_alpha = "".join([c_alpha, i])
    elif i.isdigit():
        c_num = "".join([c_num, i])
    
print("Result:",c_alpha)
print("Result:",c_num)
```

```python
from sklearn.preprocessing import LabelEncoder 
def Treatment_2(data):
    lb_make = LabelEncoder()      #Let's transform our string categorical data to numeric categorical data !    
    label_list = data.columns[5:7].tolist() #récupération des index des
    wa = data.columns[7]
    numeric_list = data.columns[~data.columns.isin(data.columns[5:8])].tolist()
    for x in data.columns: 
        for w in data.index:
            if x in numeric_list:
                if data[x][w] == 'unknown':
                    data[x][w] = 0
            elif x in wa:
                data[x][w] = data[x][w][0]
            elif x in label_list:
                data[str(x)+'_encoded'] = lb_make.fit_transform(data[x])
    if 'market_share' in data.columns.tolist():
        data = data.drop('market_share', axis = 1)
    data = data.drop(label_list, axis = 1)
    for x in data.columns:
        data[x] = data[x].astype(float)    
    return data
Treatment_2(data)
```
